encoding the target class

```python
letter2 = pd.get_dummies(letter1,columns = ['Class'],drop_first=True)
letter2.head()
```

```python
letter2.rename(columns ={'Class_G':'Target'}, inplace=True)
letter2.columns
```

The data doesnt need to be scaled because the mean and the std are relatively close

```python
X = letter2.iloc[:,0:16]
X.head()
```

```python
y = letter2[['Target']]
y.head()
```

encoding the target variable

```python
letter2.Target.value_counts()
```

Test_Train Split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.4, random_state=7)
```

defining the base_estimator with max_dept = 1

```python
from sklearn.tree import DecisionTreeClassifier
treeClf = DecisionTreeClassifier(max_depth = 1, random_state = 7)
```

creating the ADABoost with varying number of weak learners from 1 to 16

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn import metrics
Est_range = range(1,17)
score = []
for i in Est_range:
    abc = AdaBoostClassifier(n_estimators=i, base_estimator=treeClf,learning_rate=1)
    abc.fit(X_train,y_train)
    y_pred = abc.predict(X_test)
    score.append(metrics.accuracy_score(y_test,y_pred))
dict(zip(Est_range,score))
```

Now repeating the model with max depth =2

```python
from sklearn.tree import DecisionTreeClassifier
treeClf = DecisionTreeClassifier(max_depth = 2, random_state = 7)
from sklearn.ensemble import AdaBoostClassifier
from sklearn import metrics
Est_range = range(1,17)
score = []
for i in Est_range:
    abc = AdaBoostClassifier(n_estimators=i, base_estimator=treeClf,learning_rate=1)
    abc.fit(X_train,y_train)
    y_pred = abc.predict(X_test)
    score.append(metrics.accuracy_score(y_test,y_pred))
dict(zip(Est_range,score))
```

```python
from sklearn import metrics
cm = metrics.confusion_matrix(y_test,y_pred)
cm
```
