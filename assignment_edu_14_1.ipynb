Feature Selection - approach 2: Feature Importance using ExtraTreesClassifier

```python
from sklearn.ensemble import ExtraTreesClassifier
import matplotlib.pyplot as plt
model = ExtraTreesClassifier()
X = df.iloc[:,0:41]
y = df[['target']]
model.fit(X,y)
print(model.feature_importances_)
feature_imp = pd.DataFrame({'ETC_score':model.feature_importances_})
feature_imp
dfcolumns = pd.DataFrame({'features':X.columns})
dfcolumns
featureimp = pd.concat([dfcolumns,feature_imp],axis=1)
featureimp.set_index('features',inplace=True)
featureimp_sorted = featureimp.sort_values('ETC_score')
featureimp_sorted.plot(kind='barh')
plt.show()
featureimp
```

Feature Selection - approach 3: Feature selection using RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()
clf.fit(X,y)
print(clf.feature_importances_)
impfeature = pd.DataFrame({'RFC_score':clf.feature_importances_})
dfcolumns = pd.DataFrame({'features':X.columns})
dfcolumns
impfeature = pd.concat([dfcolumns,impfeature],axis=1)
impfeature.set_index('features',inplace=True)
impfeature
impfeature.plot(kind='barh')
plt.show()
RFC = featureimp
RFC
```

Feature Selection - approach 4:Recursive Feature Elemination

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression

mod = LogisticRegression()
rfe = RFE(mod, 20)
rfe_fit = rfe.fit(X, y)
rfe.support_
featureRFE = pd.DataFrame({'RFE_score':rfe.support_})
featureRFE
RFE = pd.concat([dfcolumns,featureRFE],axis=1)
RFE = RFE.set_index('features')
RFE 
```

Combining the results of all the feature selection models used

```python
from functools import reduce
fs = [RFC,RFE,featureimp]
final_results = reduce(lambda x,y: pd.merge(x,y,on ='features'),fs)
final_results

final_results['RFE_score'] = final_results['RFE_score'].astype(int)
final_results
final_results['total_score'] = final_results['RFC_score']+final_results['RFE_score']+final_results['ETC_score']
final_results = final_results.sort_values('total_score',ascending = False)
final_results
```

getting the column numbers with the top scores

```python
final = final_results.reset_index()
col = final[final['total_score']>1]['features']
col.values
```

Multicollinearity - Variable Inflation Factor

```python
df.columns
```

```python
import statsmodels.api as sm
Rsquare=[]
VIF = []
for i in range(0,41):
    y = df.iloc[:,df.columns == i]
    X = df.iloc[:,df.columns != i]
    model = sm.OLS(y,X)
    results = model.fit()
    rsq = results.rsquared
    Rsquare.append(rsq)
    vif = round(1/(1-rsq),2)
    VIF.append(vif)
    print('R Square value of {} column is {} keeping all the value as features'.format(i,(round(rsq,2))))
    print('VIF of {} column is {} \n'.format(i,vif))
```

```python
mul_col = pd.DataFrame({'Rsquare':Rsquare, 'VIF':VIF})
mul_col.sort_values('VIF', ascending = False)
```

```python
x = mul_col[mul_col['VIF']>100].index
x.values
```

```python
Using feature results and the results from multi col, we will now list the X
```

```python
# these features are to be excluded
A = set(col.values)
B = set(x)
C = A.intersection(B)
C
```

```python
# final set of features
Featureset = list(A.difference(C))
Featureset
```

```python
X = df.iloc[:,Featureset]
X.head()
```

```python
y = df[['target']]
y.head()
```

Scaling the data

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

Test_Train split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_scaled,y,test_size=0.3, random_state=7)
```

Model 1: Using Logistic_Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
logreg=LogisticRegression()
logreg.fit(X_train,y_train)
y_pred = logreg.predict(X_test)
print(metrics.accuracy_score(y_test,y_pred))
```

Model 2: Using KNN with hyper-parameter tuning

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
k_range=range(1,26)
score = []
for k in k_range:
    X_train,X_test,y_train,y_test=train_test_split(X_scaled,y,test_size=0.3, random_state=7)
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train,y_train)
    y_pred = knn.predict(X_test)
    score.append(metrics.accuracy_score(y_test,y_pred))
#thelist prints all the value of accuracy for each K
score    
dict(zip(k_range,score))
```

```python
#now plotting the relationship between K and testing accuracy

import matplotlib.pyplot as plt
plt.plot(k_range,score)
plt.xlabel('Value of K')
plt.ylabel('Accuracy')
plt.show()
```

Model 4: Using Decision Tree-with hyper-parameter tuning

```python
X_train,X_test,y_train,y_test=train_test_split(X_scaled,y,test_size=0.3, random_state=7)
from sklearn.tree import DecisionTreeClassifier
treeClf = DecisionTreeClassifier(max_depth = 3, random_state = 7)
treeClf.fit(X_train,y_train)
y_pred = treeClf.predict(X_test)
print(metrics.accuracy_score(y_test,y_pred))
```

Model 4: Using Random Forest with hyper-parameter tuning

```python
from sklearn.ensemble import RandomForestClassifier
Est_range = range(1,100)
score = []
for i in Est_range:
    model=RandomForestClassifier(n_estimators = i)
    model.fit(X_train,y_train)
    y_pred = model.predict(X_test)
    score.append(metrics.accuracy_score(y_test,y_pred))
dict(zip(Est_range,score))
```

```python
import matplotlib.pyplot as plt
plt.plot(Est_range,score)
plt.xlabel('n_estimator_value')
plt.ylabel('score')
plt.show()  
```

```python
#best model with hyper parameter tuned
model_rf=RandomForestClassifier(n_estimators = 56)
model_rf.fit(X_train,y_train)
y_pred = model_rf.predict(X_test)
score.append(metrics.accuracy_score(y_test,y_pred))
```

```python
from sklearn import metrics
cm = metrics.confusion_matrix(y_test,y_pred)
cm
```
